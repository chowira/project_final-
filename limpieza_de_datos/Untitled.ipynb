```python
# Conectar a DuckDB (crea un archivo de base de datos si no existe)
con = duckdb.connect("my_database.db")

# Crear la tabla si no existe
con.execute("""
    CREATE TABLE IF NOT EXISTS my_table (
        idusuarios INTEGER,         -- Cambié "id" por "idusuarios"
        registered TIMESTAMP,
        city VARCHAR,
        country VARCHAR,
        fecha DATE
    )
""")

# Lista de archivos CSV a procesar
csv_files = [
    '../data/Usuarios/Users-2024-12-01.csv',
    '../data/Usuarios/Users-2024-12-02.csv',
    '../data/Usuarios/Users-2024-12-03.csv',
    '../data/Usuarios/Users-2024-12-04.csv',
    '../data/Usuarios/Users-2024-12-05.csv',
    '../data/Usuarios/Users-2024-12-06.csv',
    '../data/Usuarios/Users-2024-12-07.csv'
    
]

# Procesar cada archivo CSV
for csv_file in csv_files:
    # Extraer la fecha del nombre del archivo (e.g., "2024-12-01")
    fecha = '-'.join(csv_file.split('/')[-1].split('-')[1:]).replace('.csv', '')

    # Insertar datos en la tabla con la fecha añadida y omitir filas problemáticas
    con.execute(f"""
        INSERT INTO my_table 
        SELECT id AS idusuarios, registered, city, country, '{fecha}'::DATE AS fecha
        FROM read_csv_auto('{csv_file}', strict_mode=false, ignore_errors=true)
    """)

# Consultar la tabla para mostrar los datos insertados
result = con.execute("SELECT * FROM my_table").fetchall()
for row in result:
    print(row)

```

```python
%%time
# Conectar a DuckDB (crea un archivo de base de datos si no existe)
con = duckdb.connect("my_database.db")

# Crear la tabla si no existe
con.execute("""
    CREATE TABLE IF NOT EXISTS my_table (
        idusuarios INTEGER,
        registered TIMESTAMP,
        city VARCHAR,
        country VARCHAR,
        fecha DATE
    )
""")

csv_files = [
    '../data/Usuarios/Users-2024-12-01.csv',
    '../data/Usuarios/Users-2024-12-02.csv',
    '../data/Usuarios/Users-2024-12-03.csv',
    '../data/Usuarios/Users-2024-12-04.csv',
    '../data/Usuarios/Users-2024-12-05.csv',
    '../data/Usuarios/Users-2024-12-06.csv',
    '../data/Usuarios/Users-2024-12-07.csv'
    
]    
    

# Procesar cada archivo CSV
for csv_file in csv_files:
    # Extraer la fecha del nombre del archivo (e.g., "2024-12-01")
    fecha = '-'.join(csv_file.split('/')[-1].split('-')[1:]).replace('.csv', '')

    # Insertar datos en la tabla con la fecha añadida y omitir filas problemáticas
    con.execute(f"""
        INSERT INTO my_table 
        SELECT id AS idusuarios, registered, city, country, '{fecha}'::DATE AS fecha
        FROM read_csv_auto('{csv_file}', strict_mode=false, ignore_errors=true)
    """)

# Consultar la tabla y convertirla en un DataFrame
df = con.execute("SELECT * FROM my_table").fetchdf()

# Mostrar el DataFrame
print(df)
```

```python
%%time



import duckdb

# Connect to DuckDB
con = duckdb.connect("my_database.db")

# Create the table if it doesn't exist
con.execute("""
    CREATE TABLE  my_table_3 (
        channel VARCHAR,
        origin VARCHAR, -- "from" mapped as "origin"
        duration INTEGER,
        server VARCHAR,
        type VARCHAR,
        device VARCHAR,
        idusuarios VARCHAR -- Ensure this column matches the data in the CSV
    )
""")

# List of CSV files to process
csv_files_2 = [
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-01.csv',
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-02.csv',
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-03.csv',
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-04.csv',
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-05.csv',
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-06.csv',
    '../data/tabala_de_hechos/Diciembre/Diciembre/channelUsage-2024-12-07.csv'
]

# Process each CSV file
for csv_file_2 in csv_files_2:
    con.execute(f"""
        INSERT INTO my_table_3 
        SELECT 
            channel,
            "from" AS origin,
            duration,
            server,
            type,
            device,
            user AS idusuarios -- Provide a default value for "user"
        FROM read_csv_auto('{csv_file_2}', strict_mode=false, ignore_errors=true)
    """)

# Query the table and convert it to a DataFrame
df = con.execute("SELECT * FROM my_table_3").fetchdf()

# Display the DataFrame
print(df)



```
